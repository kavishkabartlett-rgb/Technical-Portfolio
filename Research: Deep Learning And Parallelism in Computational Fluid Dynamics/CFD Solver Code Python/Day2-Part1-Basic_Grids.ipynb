```python
import numpy as np
import matplotlib.pylab as plt
```

### Uniform grid distribution
In previous section, `x` was a uniformly distributed grid

```python
nx=20
x=np.linspace(0,1,nx) #Generates 20 evenly spaced values from 0 to 1 
```

```python
fig, ax=plt.subplots(1,2,figsize=(10,3)) #Creates 1 row, 2 cols plots with the total size of both plots being 10 in x 3 in

# left: how x changes 
ax[0].plot(x)

#right: plot the points, assume y to be zero everywhere
y=np.zeros_like(x)
ax[1].plot(x,y,'o') #Means plot each point as circle marker with no connecting line

plt.tight_layout()
```

### Grid Clustering
Let's cluster the grid near `x=0` and `x=1` using hyperbolic distribution

$$x=\frac{\tanh(\alpha \xi)}{\tanh(\alpha)}, \qquad \text{where} \quad \xi=\text{Uniform}[-1,1]$$

```python
nx=20
beta=2.

xi=np.linspace(-1,1,nx) #Generate 20 points between -1 and 1 
x =np.tanh(beta*xi)/np.tanh(beta)
y=np.zeros_like(x)

fig, ax=plt.subplots(1,2,figsize=(10,3))

# left: how x and xi change 
ax[0].plot(xi)
ax[0].plot(x)

#right: plot the points for x and xi
ax[1].plot(xi,y,'o')
ax[1].plot(x,y,'o')

plt.tight_layout()
```

On the right plot: it's a bit hard to see, Let's move the clustered grid to the top of the uniform one and then remove the y-axis label

```python
nx=20
beta=2.

xi=np.linspace(-1,1,nx)
x =np.tanh(beta*xi)/np.tanh(beta)
y=np.zeros_like(x)

fig, ax=plt.subplots(1,2,figsize=(10,3))

# left: how x and xi change 
ax[0].plot(xi)
ax[0].plot(x)

#right: plot the points for x and xi
ax[1].plot(xi,y,'o')
ax[1].plot(x,y+0.01,'o')
ax[1].set_yticks([])

plt.tight_layout()
```

### Scaling the grid
**Note:** The original `x` was in $[0,1]$, but now `x` is in $[-1,1]$. How do we fix this?

```python
x=(x+1.)/2.0

fig, ax=plt.subplots(1,2,figsize=(10,3))

# left: how x and xi change 
ax[0].plot(xi)
ax[0].plot(x)

#right: plot the points for x and xi
ax[1].plot(xi,y,'o')
ax[1].plot(x,y+0.01,'o')
ax[1].set_yticks([])

plt.tight_layout()
```

## 2D grid generation

```python
nx=20
x=np.linspace(0,1,nx)

ny=30
y=np.linspace(0,1,ny)
```

```python
X,Y=np.meshgrid(x,y)
```

```python
X.shape
```

```python
X=X.T
Y=Y.T
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.plot(X,Y,'o',ms=2,color='black');
```

```python
plt.plot(X,Y,color='black');
```

**Note**: This plots specific columns of X and Y and then moves to the next column. Essentially, \
`for i in range(ny): plot(X[:,i],Y[:,i])` \
At each column, x changes from 0 to 1 and y is constant \
**What if we want to plot the vertical lines too?**

Answer: Plot the transpose of the matrices. This essentially means: \
`for i in range(nx): plot(X[i,:],Y[i,:])`

```python
plt.plot(X,Y,color='black');
plt.plot(X.T,Y.T,color='red');
```
