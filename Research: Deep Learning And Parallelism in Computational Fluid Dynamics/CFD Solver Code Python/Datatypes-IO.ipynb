## Lists
It is an ordered collection of items that can be of any type. Lists are mutable, meaning their elements can be changed after the list is created. This makes them very useful for tasks that require the storage of an ordered collection of items that may need to be modified or accessed in various ways.


### Creating Lists

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PI_vectorized(n,seed=8):
    #seed
    np.random.seed(seed)
    
    # draw random numbers?
    x = np.random.uniform(0, 1, n)
    y = np.random.uniform(0, 1, n)

    r_vec=np.sqrt((x-0.5)**2+(y-0.5)**2)
    num_inside_vec=np.count_nonzero(r_vec<0.5)
    
    return 4.0*(num_inside_vec/n)
```

```python
PI_value=[]
for i in range(1,51):
    PI_value.append(PI_vectorized(100000*i,seed=8))
```

```python
print(PI_value)
print(len(PI_value))
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,4))
ax.plot(PI_value)
ax.axhline(y=np.pi, color='r', linestyle='--')
ax.set_title("Estimation of PI")
ax.set_xlabel("Number of random drawings")
ax.set_ylabel("Estimated value of PI")
```

### List Methods
- Add the value of PI calculated with 100,000,000 random numbers to the end of this list.
- Create a new list with the last 5 elements of the `PI_values` list.
- Add "PI_vectorized(100000 * 50,seed=8)" and "PI_vectorized(100000 * 51,seed=8)" to the end of this new list using `extend` command.
- Add numbers "PI_vectorized(1000,seed=8)" and "PI_vectorized(10000,seed=8)" to the beginning of this new list using `insert` command. 

```python
# calculate PI with 1e8 random drawings
PI_value.append(PI_vectorized(100000000))
print(PI_value)

# Create a new list with the last 5 elements of the `PI_values` list
new_list = PI_value[-5:]
print('new_list=',new_list)

# add PI_vectorized(100000*50,seed=8) and PI_vectorized(100000*51,seed=8) to the end of the new list
value1= PI_vectorized(100000*50,seed=8)
value2= PI_vectorized(100000*51,seed=8)
new_value = [value1,value2]
print('new_value=',new_value)
new_list.extend(new_value)
print('new_list=',new_list)

# add PI_vectorized(1000,seed=8) and PI_vectorized(10000,seed=8) to the beginning of the new list using the insert command. 
new_list.insert(1, PI_vectorized(10000,seed=8))
print('new_list=',new_list)
```

- Remove the last element from the list and print it using `pop` command.

```python
# remove the last element from the new list
print('new_list.pop()=',new_list.pop())

# print the new list
print('new_list=',new_list)
```

## Advanced Function Definitions

### Variable-Length Arguments (`*args`, `**kwargs`)
Explanation
Function Definition:
plot_yxt(y, x, t, filename, *args, **kwargs): This function signature now includes *args and **kwargs to accept a variable number of additional positional and keyword arguments.
*args could be used for arguments that are passed directly to plt.subplots, such as nrows, ncols, and positional parameters like figsize.
**kwargs allows for passing further keyword arguments to plt.subplots or other functions inside plot_yxt.

Change the dpi=600 and facebolor to red using *args and **kwargs


```python
#Define the constants
g=9.81;
v_0y=5.0;
v_0x=5.0;
y0=0.;
x0=0.;

#Define the array of time
t=np.linspace(0,2,10000)

#calculate y and x
y=(-1./2.)*g*t**2+v_0y*t+y0;
x=v_0x*t+x0;

def plot_yxt(y,x,t,filename):
    fig, ax=plt.subplots(1,2,figsize=(10,4), dpi=300, subplot_kw={'facecolor': 'w'} )
    #plot the two graphs
    ax[0].plot(t,y);
    ax[1].plot(t,x);
    
    #labels for subplot 0
    ax[0].set_xlabel('t')
    ax[0].set_ylabel('y')
    
    #labels for subplot 1
    ax[1].set_xlabel('t')
    ax[1].set_ylabel('x')

    #adjust the margins
    plt.tight_layout()
    
    #save in a pdf
    plt.savefig(filename)
    
plot_yxt(y,x,t,'ballFlight_yxt.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constants
g = 9.81
v_0y = 5.0
v_0x = 5.0
y0 = 0.
x0 = 0.

# Define the array of time
t = np.linspace(0, 2, 10000)

# Calculate y and x
y = (-1. / 2.) * g * t**2 + v_0y * t + y0
x = v_0x * t + x0

def plot_yxt(y, x, t, filename, *args, **kwargs):
    fig, ax = plt.subplots(1, 2, figsize=(10, 4), *args, **kwargs)

    # Plot the two graphs
    ax[0].plot(t, y)
    ax[1].plot(t, x)

    # Labels for subplot 0
    ax[0].set_xlabel('t')
    ax[0].set_ylabel('y')

    # Labels for subplot 1
    ax[1].set_xlabel('t')
    ax[1].set_ylabel('x')

    # Adjust the margins
    plt.tight_layout()

    # Save in a pdf
    plt.savefig(filename)

# Example of calling plot_yxt with additional arguments and keyword arguments
plot_yxt(y, x, t,  filename='ballFlight_yxt.pdf', dpi=1000, subplot_kw={'facecolor': 'gray'})
```
