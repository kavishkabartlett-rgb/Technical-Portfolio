# Numerical Integration Using the Trapezoidal Rule

In this notebook, we will calculate the definite integral of the function $f(x) = x^2$ over the interval $[0, 1]$ using the **trapezoidal rule**. We'll also explain each part of the implementation step-by-step with equations and a plot.

## ðŸ“˜ What is the Trapezoidal Rule?

The trapezoidal rule approximates the area under a curve by dividing it into trapezoids rather than rectangles. For a function $f(x)$ over an interval $[a, b]$, the integral is approximated as:

\$
\int_a^b f(x)\, dx \approx \frac{h}{2} \left[f(x_0) + 2 \sum_{i=1}^{n-1} f(x_i) + f(x_n)\right]
\$

Where:
- $h = \frac{b - a}{n}$ is the width of each subinterval.
- $x_0, x_1, ..., x_n$ are the evenly spaced points between $a$ and $b$.


## Step 1: Import Libraries and Define the Function

We will use NumPy for numerical operations and Matplotlib to visualize the function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^2
def f(x):
    return x**2
```

## Step 2: Visualize the Function

Let's visualize $f(x) = x^2$ over the interval $[0, 1]$.

```python
# Create x values for plotting
x_vals = np.linspace(0, 1, 100)
y_vals = f(x_vals)

plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, label='f(x) = x^2', color='blue')
plt.title("Function f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## Step 3: Implement the Trapezoidal Rule

Now we'll compute the integral of $f(x) = x^2$ from 0 to 1 using the trapezoidal rule.

```python
# Integration bounds
a = 0
b = 1

# Number of subintervals
n = 1000

# Width of each subinterval
h = (b - a) / n

# x values
x = np.linspace(a, b, n+1)

# Apply trapezoidal rule
integral = (h/2) * (f(x[0]) + 2 * np.sum(f(x[1:n])) + f(x[n]))

print(f"Estimated integral of f(x) = x^2 from 0 to 1 using trapezoidal rule: {integral:.6f}")
```

## âœ… Conclusion

The trapezoidal rule gives us a numerical approximation of the definite integral. For $f(x) = x^2$ over $[0, 1]$, the exact value is:

\$
\int_0^1 x^2 dx = \frac{1}{3} \approx 0.333333
\$

With a sufficient number of subintervals (e.g., $n = 100$), the trapezoidal rule provides a very accurate result.
